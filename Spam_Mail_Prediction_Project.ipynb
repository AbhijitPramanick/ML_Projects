# **Spam Mail Prediction Using Machine Learning with Python**

**Work Flow**

Mail Data ---> Data Pre Processing ---> Train Test Split ---> **Logistic Regression Model**

New Mail ---> **Trained Logistic Regression Model** ---> Prediction (Spam or Ham)

**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection and Pre-processing**

```python
raw_mail_data = pd.read_csv('mail_data.csv')
```

```python
# Getting the first 5 rows of the dataset
raw_mail_data.head()
```

```python
# Getting the dimensions of the DataFrame
raw_mail_data.shape
```

```python
# Getting information about the DataFrame
raw_mail_data.info()
```

```python
# Further checking for null values if any:
raw_mail_data.isnull().sum()
```

```python
# Replacing the null values (if any) with null string
mail_data = raw_mail_data.where(pd.notnull(raw_mail_data), '')
```

**Label Encoding**

```python
# Label Spam mail as 0 and Ham mail as 1

mail_data.loc[mail_data['Category']=='spam', 'Category'] = 0
mail_data.loc[mail_data['Category']=='ham', 'Category'] = 1
```

Spam -> 0

Ham -> 1

```python
# Seperating the texts and label

X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the data into Training and Testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2, stratify = Y)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

**Feature Extraction**

```python
# Transform the text data into numeric Data (Feature vector) that can be used into Logistic Regression Model

feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True,)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# Convert Y_train and Y_test data into integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
# Visualising the X_train_features
print(X_train_features)
```

**Training the Logistic Regression Model**

```python
model = LogisticRegression()
model.fit(X_train_features, Y_train)
```

```python
X_train_prediction = model.predict(X_train_features)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy of the model's prediction on Training Data: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test_features)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy of the model's prediction on Testing Data: ", testing_data_accuracy)
```
